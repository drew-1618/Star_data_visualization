```python
# importing libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# preview data
star_df = pd.read_csv("Data/cleaned_star_data.csv")
star_df.sample(10)
```

```python
# general info about the data
star_df.info()
# Star Type numerical associations:
    # 0 → Brown Dwarf
    # 1 → Red Dwarf
    # 2 → White Dwarf
    # 3 → Main Sequence
    # 4 → Supergiants
    # 5 → Hypergiants
```

```python
# create directory to save visualizations
folder_name = 'star_plots'
os.makedirs(folder_name, exist_ok=True)  
base_dir = f'/users/andrewjohnson/PersonalProjects/AstronomicalDataAnalysis/{folder_name}/'
```

```python
# customized bar chart to visualize count of stars with respective types
plt.figure(figsize=(4.5, 4.5))
# must run twice to get both bar background axis background changed
plt.style.use('dark_background')
colors = ['#4E79A7', '#F28E2B', '#76B7B2', '#A0A0A0', '#E1C35A', '#B07AA1']
ax = star_df['Star type'].value_counts().plot(kind='bar', color=colors)
ax.bar_label(ax.containers[0], color='#F0F0F0')
plt.title('Visualize Star Count per Star Type', color='royalblue', weight='bold')
plt.ylabel('# of Stars', fontsize=12, color='orange')
plt.xlabel('Star type', fontsize=12, color='orange')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], 
           labels=['Brown\nDwarf', 'Red\nDwarf', 'White\nDwarf', 'Main\nSequence', 'Supergiant', 'Hypergiant'],
           rotation=45)
# prevent cut off
plt.tight_layout()
plt.savefig(base_dir + 'barplot_star_count_by_type.png')
plt.show()
```

```python
# bar chart to visualize star color data using Seaborn
ax = sns.barplot(x=star_df['Star color'].value_counts().index,
            y=star_df['Star color'].value_counts(), 
            hue=star_df['Star color'].value_counts().index,
            palette='viridis')
plt.ylabel('Count', fontsize=12, color='orange')
plt.xlabel('Star color', fontsize=12, color='orange')
plt.title('Visualize Star Count per Color', color='royalblue', weight='bold')
ax.bar_label(ax.containers[0], color='#F0F0F0')
plt.savefig(base_dir + 'barplot_star_count_by_color.png')
plt.show()
```

```python
# visualize outliers in the data
plt.figure(figsize=(20, 8))
plt.style.use('default')
plt.suptitle('Visualizing Outliers in Numeric Features of Star Type', weight='bold', color='black', fontsize=20)
for i in range(4):
    plt.subplot(1, 4, i+1)
    sns.boxplot(x=star_df['Star type'], y=star_df.iloc[:, i])   # column index
    plt.ylabel('')
    plt.title(star_df.columns[i], color='red')
plt.savefig(base_dir + 'boxplot_star_type.png')
plt.show()
```

```python
def line_subplot(i, star_df, colors):
    plt.subplot(4, 1, i+1)
    plt.plot(star_df.iloc[:,i], color=colors[i])
    plt.title(f'Distribution of {star_df.columns[i]}', color='red')
```

```python
# create line plots
colors = ['royalblue', 'magenta', 'lime', 'cyan']
plt.figure(figsize=(12,8))
plt.suptitle('Visualize Distribution of Numeric Features', color='black', weight='bold', fontsize=15)
for i in range(4):
    line_subplot(i, star_df, colors)
plt.tight_layout()
plt.savefig(base_dir + 'line_subplot_distributions.png')
plt.show()
```

```python
# make pairplot - scatter
sns.pairplot(star_df, hue='Spectral Class')
plt.savefig(base_dir + 'pairplot.png')
plt.show()
```

```python
star_type = star_df['Star type'].values
temperature = star_df['Temperature (K)'].values
abs_mag = star_df['Absolute magnitude(Mv)'].values

star_types = { 0: { 'label' : 'Brown Dwarf', 'color' : 'brown', 'size' : 30, 'marker' : '.' },
               1: { 'label' : 'Red Dwarf', 'color' : 'red', 'size' : 35, 'marker' : '.' },
               2: { 'label' : 'White Dwarf', 'color' : 'white', 'size' : 40, 'marker' : '.' },
               3: { 'label' : 'Main Sequence', 'color' : 'cyan', 'size' : 30, 'marker' : '.' },
               4: { 'label' : 'Supergiant', 'color' : 'orange', 'size' : 100, 'marker' : 'o' },
               5: { 'label' : 'Hypergiant', 'color' : 'maroon', 'size' : 150, 'marker' : 'o' },
             }
```

```python
# plotting HR diagram for Temperature & Absolute magnitude 
labels = set()
axes = []
transparency = 0.6
plt.figure(figsize=(10, 6))
plt.style.use('dark_background')

for i in range(len(star_type)):
    properties = star_types[star_type[i]]
    if properties['label'] not in labels:
        ax = plt.scatter(temperature[i], abs_mag[i],
                    s = properties['size'],
                    c = properties['color'],
                    marker = properties['marker'],
                    label = properties['label'], 
                    alpha=transparency
                   )
        axes.append(ax)
        labels.add(properties['label'])
    else:
        plt.scatter(temperature[i], abs_mag[i],
                    s = properties['size'],
                    c = properties['color'],
                    marker = properties['marker'],
                    label = properties['label'],
                    alpha=transparency
                   )

# Add data for sun
ax_sun = plt.scatter(5778, 4.83, s=75, c='yellow', marker='*', label='Sun')
axes.append(ax_sun)
labels.add('Sun')

plt.title(f'Hertzsprung - Russell Diagram for {len(star_type)} Stars', color='royalblue', fontsize=15, weight='bold')
plt.xlabel('Temperature (K)', color='orange', fontsize=13)
plt.ylabel('Absolute Magnitude (Mv)', color='orange', fontsize=13)

plt.legend(handles=axes)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.savefig(base_dir + 'HR_diagram.png')
plt.show()
```
